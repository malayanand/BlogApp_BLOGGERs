```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("2021preprocessed.csv")
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1500, stop_words='english')
# bag-of-words feature matrix
bow = bow_vectorizer.fit_transform(df['Tweets'])
df_bow = pd.DataFrame(bow.todense())
df_bow
```

```python
tfidf=TfidfVectorizer(max_df=0.90, min_df=2,max_features=1500,stop_words='english')
tfidf_matrix=tfidf.fit_transform(df['Tweets'])
df_tfidf = pd.DataFrame(tfidf_matrix.todense())
df_tfidf
```

```python
train_bow = bow[:]
train_bow.todense()
```

```python
train_tfidf_matrix = tfidf_matrix[:]
train_tfidf_matrix.todense()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# bag of words
x_train_bow, x_valid_bow, y_train_bow, y_valid_bow = train_test_split(train_bow,df['Classification'],test_size=0.3,random_state=2)
# tfidf
x_train_tfidf, x_valid_tfidf, y_train_tfidf, y_valid_tfidf = train_test_split(train_tfidf_matrix,df['Classification'],test_size=0.3,random_state=17)
```

```python
from sklearn.linear_model import LogisticRegression
Log_Reg = LogisticRegression(random_state=0,solver='lbfgs')
```

**Bag of words feature set**

```python
Log_Reg.fit(x_train_bow,y_train_bow)
```

```python
prediction_bow = Log_Reg.predict_proba(x_valid_bow)
prediction_bow
```

```python
prediction = []

for each_prob in prediction_bow:
  arr = np.array(each_prob)
  max_val = max(each_prob)
  max_index = np.where(arr == max_val)

  if max_index[0][0] == 0:
    prediction.append(-1)
  elif max_index[0][0] == 1:
    prediction.append(0)
  elif max_index[0][0] == 2:
    prediction.append(1)


print(prediction)
print(len(prediction))
```

```python
from collections import Counter
counts_bow = dict(Counter(prediction))
counts_bow
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,6))
plt.title('Sentiment Analysis: Bag of words')
plt.xlabel('Sentiments')
plt.ylabel('Counts')

sentiment = list(counts_bow.keys())
count = list(counts_bow.values())

plt.bar(sentiment, count, color ='maroon', width = 0.4)
plt.show()
```

```python
mylabels = ["Neutral", "Positive", "Negative"]
plt.figure(figsize=(8,6))
plt.title('Sentiment Analysis: Bag of words Pie plot')
plt.pie(count, labels=mylabels)
plt.show()
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

print("Precision Score : ", precision_score(y_valid_bow, prediction, average='micro'))
print("Recall Score : ", recall_score(y_valid_bow, prediction, average='micro'))
print("F1 score: ", f1_score(y_valid_bow, prediction, average='weighted'))
```

**Tfidf feature set**

```python
Log_Reg.fit(x_train_tfidf,y_train_tfidf)
prediction_tfidf = Log_Reg.predict_proba(x_valid_tfidf)
prediction_tfidf
```

```python
prediction_with_tfidf = []

for each_prob in prediction_tfidf:
  arr = np.array(each_prob)
  max_val = max(each_prob)
  max_index = np.where(arr == max_val)

  if max_index[0][0] == 0:
    prediction_with_tfidf.append(-1)
  elif max_index[0][0] == 1:
    prediction_with_tfidf.append(0)
  elif max_index[0][0] == 2:
    prediction_with_tfidf.append(1)


print(prediction_with_tfidf)
```

```python
print("Precision Score : ", precision_score(y_valid_tfidf, prediction_with_tfidf, average='micro'))
print("Recall Score : ", recall_score(y_valid_tfidf, prediction_with_tfidf, average='micro'))
print("F1 score: ", f1_score(y_valid_tfidf, prediction_with_tfidf, average='weighted'))
```

```python
from collections import Counter
counts_tfidf = dict(Counter(prediction_with_tfidf))
counts_tfidf
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,6))
plt.title('Sentiment Analysis: Tfidf')
plt.xlabel('Sentiments')
plt.ylabel('Counts')

sentiment = list(counts_tfidf.keys())
count = list(counts_tfidf.values())

plt.bar(sentiment, count, color ='maroon', width = 0.4)
plt.show()
```

```python
mylabels = ["Negative", "Neutral", "Positive"]
myexplode = [0.2, 0, 0]
plt.figure(figsize=(8,6))
# plt.title('Sentiment Analysis: Tfidf Pie plot')
plt.pie(count, labels=mylabels, explode=myexplode)
plt.show()
```
